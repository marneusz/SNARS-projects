```python
seasons = range(1, 9)

Gs = []
for season in seasons:
    s = pd.read_csv(os.path.join(path_to_data, f'got-s{season}-edges.csv'))
    G_s = nx.Graph()
    for _, edge in s.iterrows():
        G_s.add_edge(edge['Source'], edge['Target'], weight=edge['Weight'])
    Gs.append(G_s)
```

```python
season = 1
G = Gs[season-1]
G2 = Gs[2-1]
```

```python
print(nx.info(G))
```

```python
plt.figure(figsize = (10,10))
nx.draw(G, pos = nx.spring_layout(G), \
    node_size=70, node_color='darkblue', alpha = 0.4)
```

## Degree Distribution

```python
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
fig = px.histogram(degree_sequence)
fig.update_layout(
    xaxis_title="Degree",
    yaxis_title="Count",
    title="Distribution of Degrees in the Graph for Season 1"
)
fig.show()
```

## Is the network scale-free?

```python
np.seterr(divide='ignore', invalid='ignore')

fit = powerlaw.Fit(degree_sequence) 
# x_min is calculated based on Kolmogorov-Smirnov distance
fig = fit.plot_pdf(color='b', linewidth=2, )
fit.power_law.plot_pdf(color='g', linestyle='--', ax=fig)
plt.xlabel("Index")
plt.ylabel("Degree")
plt.title('Comparison for optimal xmin')
plt.show()

fit2 = powerlaw.Fit(degree_sequence, xmin=1) 
fig2 = fit2.plot_pdf(color='b', linewidth=2, )
fit2.power_law.plot_pdf(color='g', linestyle='--', ax=fig2)
plt.xlabel("Index")
plt.ylabel("Degree")
plt.title('Comparison for xmin equal to 1')
plt.show()
```

# Game of Thrones - Season 8

```python
season = 8
edges = pd.read_csv(os.path.join(path_to_data, f'got-s{season}-edges.csv'))
got_net = Network(height='750px', width='100%', bgcolor='#222222', font_color='white')
got_net.barnes_hut()


sources = edges['Source']
targets = edges['Target']
weights = edges['Weight']
edge_data = zip(sources, targets, weights)

for e in edge_data:
    src = e[0]
    dst = e[1]
    w = e[2]

    got_net.add_node(src, src, title=src)
    got_net.add_node(dst, dst, title=dst)
    got_net.add_edge(src, dst, value=w)
```

```python
neighbor_map = got_net.get_adj_list()

# add neighbor data to node hover data
for node in got_net.nodes:
    node['title'] += ' Neighbors:<br>' + '<br>'.join(neighbor_map[node['id']])
    node['value'] = len(neighbor_map[node['id']])

got_net.show('gameofthrones.html')
```

# Bitcoin OTC Trust Weighted Signed Network

Source: [Bitcoin OTC trust weighted signed network](http://snap.stanford.edu/data/soc-sign-bitcoin-otc.html)

This is who-trusts-whom network of people who trade using Bitcoin on a platform called Bitcoin OTC. Since Bitcoin users are anonymous, there is a need to maintain a record of users' reputation to prevent transactions with fraudulent and risky users. Members of Bitcoin OTC rate other members in a scale of -10 (total distrust) to +10 (total trust) in steps of 1. This is the first explicit weighted signed directed network available for research.

```python
df = pd.read_csv("/mnt/c/Users/01130781/OneDrive - Politechnika Warszawska/Pulpit/Zajęcia/SNARS/Data/soc-sign-bitcoinotc.csv", 
                 header=None, names=["SOURCE", "TARGET", "RATING", "TIME"])
df.head()
```

```python
G_s = nx.Graph()
for _, edge in df.iterrows():
    G_s.add_edge(int(edge['SOURCE']), int(edge['TARGET']), weight=edge['RATING'])
```

```python
nx.info(G_s)
```

```python
plt.figure(figsize = (15,15))
nx.draw(G_s, 
        pos = nx.spring_layout(G_s),
        node_size=30, 
        node_color='darkblue', 
        alpha = 0.1)
```

```python
btc_net = Network(height='750px', width='100%', bgcolor='#222222', font_color='white')
btc_net.barnes_hut()
```

```python
sources = df['SOURCE']
targets = df['TARGET']
weights = df['RATING']
edge_data = zip(sources, targets, weights)

for e in edge_data:
    src = e[0]
    dst = e[1]
    w = e[2]

    btc_net.add_node(src, src, title=str(src))
    btc_net.add_node(dst, dst, title=str(dst))
    btc_net.add_edge(src, dst, value=w)
```
