### Generating Graph with Communities

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
num_of_vertices = 10
num_of_communities = 6
```

```python
max_prob = 0.15
```

```python
between_communities = np.random.rand(num_of_vertices * num_of_communities, num_of_vertices * num_of_communities) * max_prob
```

```python
for i in range(num_of_communities):
    between_communities[(i*num_of_vertices):((i+1)*num_of_vertices), 
                        (i*num_of_vertices):((i+1)*num_of_vertices)] = (np.random.rand(num_of_vertices, num_of_vertices) + 1) / 2
```

```python
probability_matrix = (between_communities + between_communities.T)/2
np.fill_diagonal(probability_matrix, 0)
sns.heatmap(probability_matrix, cmap='coolwarm', ax=plt.subplots(figsize=(10,8))[1]) # probability matrix 
```

```python
adjacency_matrix = probability_matrix > np.random.rand(num_of_vertices*num_of_communities, num_of_vertices*num_of_communities)  
G = nx.convert_matrix.from_numpy_array(adjacency_matrix)
nx.draw_kamada_kawai(G, with_labels=False, node_size=25, alpha=0.25, ax=plt.subplots(figsize=(10,8))[1])
```

```python
nx.draw_networkx(G, nx.spring_layout(G, iterations=1000), with_labels=False, node_size=20, alpha=0.4, ax=plt.subplots(figsize=(10,8))[1])
```

### P10.1

```python
def prepare_laplacian(A: np.matrix) -> np.matrix:
    """Given an adjacency matrix, computes the laplacian matrix"""
    return A - np.diag(A.sum(axis=0))
```

```python
L = prepare_laplacian(adjacency_matrix)
L
```

```python
K = num_of_communities  # Number of Communities
```

```python
def get_features(L: np.matrix) -> np.matrix:
    """Given a Laplacian matrix, return the feature vector based on eigenvalues"""
    eigval, eigvect = np.linalg.eig(L)
    K_largest_eigvals_indices = np.argsort(-np.real(eigval))[:K]
    return eigvect[:, K_largest_eigvals_indices]
```

```python
features = pd.DataFrame(np.real(get_features(L)))
features[:10]
```

```python
sns.pairplot(features)
```

### K-Means

```python
from sklearn.cluster import k_means
```

```python
labels = k_means(features, K)[1]
labels
```

```python
features["labels"] = labels
sns.pairplot(features, hue="labels")
```

```python
nx.draw_networkx(G, nx.spring_layout(G, iterations=10000), with_labels=False, node_size=100, ax=plt.subplots(figsize=(15,10))[1], node_color=labels)
```

### High-Level Implementations in Networkx

```python
nx.__version__
```

```python
from networkx.algorithms import community
```

https://en.wikipedia.org/wiki/Girvan%E2%80%93Newman_algorithm

The algorithm's steps for community detection are summarized below

1. The betweenness of all existing edges in the network is calculated first.
2. The edge(s) with the highest betweenness are removed.
3. The betweenness of all edges affected by the removal is recalculated.
4. Steps 2 and 3 are repeated until no edges remain.

https://medium.com/smucs/the-implementation-and-analysis-of-the-girvan-newman-algorithm-part-1-3ce999c04cb4

```python
import itertools
communities_generator = community.girvan_newman(G)
for communities in itertools.islice(communities_generator, K-1):
    print(tuple(sorted(c) for c in communities))
```
