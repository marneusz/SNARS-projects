## P1.8 Draw square lattice with your own function. 

```python
class SquareLattice(Graph):
    def __init__(self, x, y):
        self.graph = dict()
        self.x = x
        self.y = y
        
        if (self.x == 1) and (self.y == 1):
            super().add_vertex(0)
        elif self.x == 1:
            for i in range(self.y - 1):
                super().add_edge(i, i + 1)
        elif self.y == 1:
            for i in range(self.x - 1):
                super().add_edge(i, i + 1)
        else:
            for i in range(self.x):
                for j in range(self.y):
                    if i < self.x - 1:
                        super().add_edge(i + j*self.x, i + 1 + j*self.x)

                    if j < self.y - 1:
                        super().add_edge(i + j*self.x, i + (j + 1) * self.x)
                
```

```python
sl = SquareLattice(3, 5)
```

```python
sl.get_all_vertices()
```

```python
sl.get_all_edges()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(sl.get_all_edges())
```

```python
#  https://networkx.org/documentation/stable/tutorial.html#drawing-graphs
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

```python
chain_dict = {
    0: 10,
    2: 3,
    5: 2,
    7: 5,
}
```

```python
class ChainGraph(Graph):
    def __init__(self, chain_dict: dict = None):
        self.graph = dict()
        if chain_dict is None:
            self.add_vertex(0)
            return
        
        for k in chain_dict:
            num_of_vertices = len(self.get_all_vertices())
            for i in range(chain_dict[k]):
                self.add_edge((k if i == 0 else i + num_of_vertices), i + 1 + num_of_vertices, two_way=False)
    
```

```python
chain_graph = ChainGraph(chain_dict)
```

```python
G = nx.DiGraph()
G.add_edges_from(chain_graph.get_all_edges())
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

```python
nx.is_directed_acyclic_graph(G)
```

## Draw complete graph with your own function.

```python
import math
```

```python
class CompleteGraph(Graph):
    def __init__(self, n):
        super().__init__()
        if n == 1:
            self.add_vertex(0)
            return
        for i in range(n):
            for j in range(i + 1, n):
                self.add_edge(i, j)
```

```python
n = 10
complete_graph = CompleteGraph(n)
```

```python
len(complete_graph.get_all_edges()) / 2 == math.comb(n, 2)
```

```python
G = nx.Graph()
G.add_edges_from(complete_graph.get_all_edges())
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```
