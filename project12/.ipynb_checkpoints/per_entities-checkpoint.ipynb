```python
import json
import numpy as np
from tqdm.notebook import tqdm
from operator import itemgetter
import re
import pandas as pd

import networkx as nx
from matplotlib.pyplot import figure

from fuzzywuzzy import fuzz, process
```

```python
def flatten(x):
    result = []
    for el in x:
        if hasattr(el, "__iter__") and not isinstance(el, str):
            result.extend(flatten(el))
        else:
            result.append(el)
    return result
```

```python
with open("parsed_tweets.json") as f:
    parsed_tweets = json.load(f)
```

```python
tags = []

for t in tqdm(parsed_tweets):
    for e in t['entities']:
        tags.append(e['tag'])
        
tags = list(np.unique(tags))

tags
```

```python
tag = tags[2]  # param
```

```python
nested_lst = []

for t in tqdm(parsed_tweets):
    lst = []
    for e in t['entities']:
        if e['tag'] == tag:
            _s = " ".join(e['entity'])
            _s = re.sub(r'[^A-Za-z ]+', '', _s)
            _s = re.sub('\s+', ' ', _s)
            s = str.lower(_s.lstrip().rstrip())
            if len(s) > 0: lst.append(s)
    if len(lst) > 0: nested_lst.append(lst)
```

```python
# get rid of the empty lists within the list of list that were generated
# by tweets that didnâ€™t have any {tag} entity types
tweets = [l for l in nested_lst if l != []]
```

```python
G = nx.Graph()

for lst in tweets:
    G.add_nodes_from(lst)
```

```python
for lst in nested_lst:
    if len(lst) > 1:
        for i in range(len(lst)-1):
            G.add_edges_from([(str(lst[i]),str(lst[i+1]))])
```

```python
degree_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, degree_dict, 'degree')
sorted_degree = sorted(degree_dict.items(), key=itemgetter(1), reverse=True)

n = 75

print(f"Top {n} nodes by degree:")
for d in sorted_degree[:n]:
    print(d)
```

```python
pd.DataFrame(nested_lst).to_csv(f"{tag}.csv",index=False)
```

```python
outdeg = G.degree()
to_keep = [n[0] for n in outdeg if n[1] >= 400]
```

```python
G1 = G.subgraph(to_keep)
```

```python
figure(figsize=(10, 8))
nx.draw(G1, with_labels=True)
```

```python
# [n[0] for n in outdeg if n[1] > 25]
to_keep = to_keep + [
    'hillary clinton',
    'trump',
    'hillary',
    'obama',
    'clinton',
    'bill clinton',
    'donald',
    'donald trump',
    'merkel',
    'donald trumps',
    'hillary clintons',
    'putin',
    'michelle obama',
    'barack obama',
    'bill clintons',
    'obamas',
    'trumps',
    'hitler',
    'bush',
    'reagan',
    'melania trump',
    'johnson'
    'george soros',
    'soros']  # add politicians with degree > 25
```

```python
nested_lst_clean = []  # too many observations for full cleaning --> choices prunning
choices = list(np.unique(to_keep))

for lst in tqdm(tweets):
    alt_lst = []
    for e in lst:
        alt_e = process.extractOne(e, choices)
        alt_lst.append(alt_e[0])
    nested_lst_clean.append(alt_lst)
```

```python
G2 = nx.Graph()

for lst in nested_lst_clean:
    G2.add_nodes_from(lst)
```

```python
for lst in nested_lst_clean:
    if len(lst) > 1:
        for i in range(len(lst)-1):
            G2.add_edges_from([(str(lst[i]),str(lst[i+1]))])
```

```python
degree_dict = dict(G2.degree(G2.nodes()))
nx.set_node_attributes(G2, degree_dict, 'degree')
sorted_degree = sorted(degree_dict.items(), key=itemgetter(1), reverse=True)

for d in sorted_degree[:n]:
    print(d)
```

```python
figure(figsize=(10, 8))
nx.draw_shell(G2, with_labels=True)
```
