```python
import json
import numpy as np
from tqdm.notebook import tqdm
from operator import itemgetter
import re
import pandas as pd

import networkx as nx
from matplotlib.pyplot import figure

from fuzzywuzzy import fuzz, process
```

```python
def flatten(x):
    result = []
    for el in x:
        if hasattr(el, "__iter__") and not isinstance(el, str):
            result.extend(flatten(el))
        else:
            result.append(el)
    return result
```

```python
with open("parsed_tweets.json") as f:
    parsed_tweets = json.load(f)
```

```python
tags = []

for t in tqdm(parsed_tweets):
    for e in t['entities']:
        tags.append(e['tag'])
        
tags = list(np.unique(tags))

tags
```

```python
tag = tags[0]  # param
```

```python
nested_lst = []

for t in tqdm(parsed_tweets):
    lst = []
    for e in t['entities']:
        if e['tag'] == tag:
            _s = " ".join(e['entity'])
            _s = re.sub(r'[^A-Za-z ]+', '', _s)
            _s = re.sub('\s+', ' ', _s)
            s = str.lower(_s.lstrip().rstrip())
            if len(s) > 0: lst.append(s)
    if len(lst) > 0: nested_lst.append(lst)
```

```python
# get rid of the empty lists within the list of list that were generated
# by tweets that didnâ€™t have any {tag} entity types
tweets = [l for l in nested_lst if l != []]

# flatted nested list for fuzzywuzzy
all_entities = flatten(tweets)
choices = list(np.unique(all_entities))
```

```python
nested_lst_clean = []

for lst in tqdm(tweets):
    alt_lst = []
    for e in lst:
        alt_e = process.extractOne(e, choices)
        alt_lst.append(alt_e[0])
    nested_lst_clean.append(alt_lst)
```

```python
G = nx.Graph()

for lst in nested_lst_clean:
    G.add_nodes_from(lst)
```

```python
for lst in nested_lst_clean:
    if len(lst) > 1:
        for i in range(len(lst)-1):
            G.add_edges_from([(str(lst[i]),str(lst[i+1]))])
```

```python
figure(figsize=(10, 8))
nx.draw(G, node_size=10)
```

```python
figure(figsize=(10, 8))
nx.draw_shell(G, node_size=15)
```

```python
degree_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, degree_dict, 'degree')
sorted_degree = sorted(degree_dict.items(), key=itemgetter(1), reverse=True)

n = 100

print(f"Top {n} nodes by degree:")
for d in sorted_degree[:n]:
    print(d)
```

```python
pd.DataFrame(nested_lst_clean).to_csv(f"{tag}.csv",index=False)
```
